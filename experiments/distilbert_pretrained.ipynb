## Pretrained distilbert

```python
from transformers import pipeline

pipe = pipeline("text-classification", model="sarahai/movie-sentiment-analysis", truncation=True)
```

```python
reviews = ["I liked it very much", "I will never watch this"]
pipe(reviews)
```
